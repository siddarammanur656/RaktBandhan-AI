```python

# BLOOD DONOR PREDICTION SYSTEM
# Cell 1: Imports & Dataset Loading

import warnings
warnings.filterwarnings("ignore")

import pandas as pd
import numpy as np

# Display Settings
pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", 100)

# Reproducibility
RANDOM_STATE = 42

# Load Dataset
df = pd.read_csv("Dataset.csv")

print("=" * 50)
print("DATASET LOADED SUCCESSFULLY")
print("=" * 50)

print(f"Rows    : {df.shape[0]}")
print(f"Columns : {df.shape[1]}")

print("\nDataset Preview:")
display(df.head())

print("\nColumn Names:")
print(df.columns.tolist())
```

```python

# Cell 2: Filter Donors & Create Target

# Keep only donor records
donor_df = df[
    df["role"].isin([
        "Emergency Donor",
        "Bridge Donor"
    ])
].copy()

# Create Target Variable
donor_df["target"] = np.where(
    donor_df["donated_earlier"] == True,
    1,
    0
)

print("=" * 50)
print("DONOR DATASET")
print("=" * 50)

print(f"Rows    : {donor_df.shape[0]}")
print(f"Columns : {donor_df.shape[1]}")

print("\nTarget Distribution:")
print(donor_df["target"].value_counts())

print("\nTarget Percentage:")
print(
    round(
        donor_df["target"].value_counts(normalize=True) * 100,
        2
    )
)
```

```python

# Cell 3: Feature Selection

FEATURES = [
    "blood_group",
    "gender",
    "latitude",
    "longitude",
    "donor_type",
    "eligibility_status",
    "user_donation_active_status"
]

TARGET = "target"

X = donor_df[FEATURES].copy()
y = donor_df[TARGET].copy()

print("=" * 50)
print("FEATURE DATASET")
print("=" * 50)

print("X Shape:", X.shape)
print("y Shape:", y.shape)

print("\nMissing Values:")
print(X.isnull().sum())

print("\nFeature Types:")
print(X.dtypes)
```

```python

# Cell 4: Train Test Split


from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)

print("=" * 50)
print("TRAIN TEST SPLIT")
print("=" * 50)

print("X_train:", X_train.shape)
print("X_test :", X_test.shape)

print("\nTraining Target Distribution")
print(y_train.value_counts())

print("\nTesting Target Distribution")
print(y_test.value_counts())
```

```python

# Cell 5: Data Preprocessing Pipeline

from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder

# Separate feature types
categorical_features = [
    "blood_group",
    "gender",
    "donor_type",
    "eligibility_status",
    "user_donation_active_status"
]

numerical_features = [
    "latitude",
    "longitude"
]

# Numerical preprocessing
numeric_transformer = Pipeline(
    steps=[
        ("imputer", SimpleImputer(strategy="median"))
    ]
)

# Categorical preprocessing
categorical_transformer = Pipeline(
    steps=[
        ("imputer", SimpleImputer(strategy="most_frequent")),
        ("encoder", OneHotEncoder(handle_unknown="ignore"))
    ]
)

# Combine preprocessing
preprocessor = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, numerical_features),
        ("cat", categorical_transformer, categorical_features)
    ]
)

print("=" * 50)
print("PREPROCESSOR CREATED SUCCESSFULLY")
print("=" * 50)

print("Categorical Features:", len(categorical_features))
print("Numerical Features  :", len(numerical_features))
```

```python
# ==========================================================
# Cell 6: XGBoost Model Configuration
# ==========================================================

from xgboost import XGBClassifier

model = XGBClassifier(
    n_estimators=100,
    max_depth=4,
    learning_rate=0.05,
    subsample=0.8,
    colsample_bytree=0.8,
    min_child_weight=5,
    random_state=42,
    eval_metric="logloss"
)

print("=" * 50)
print("XGBOOST MODEL CREATED")
print("=" * 50)

print(model)
```

```python

# Cell 7: Complete ML Pipeline

from sklearn.pipeline import Pipeline

pipeline = Pipeline(
    steps=[
        ("preprocessor", preprocessor),
        ("classifier", model)
    ]
)

print("=" * 50)
print("PIPELINE CREATED SUCCESSFULLY")
print("=" * 50)

print(pipeline)
```

```python
pipeline.fit(X_train, y_train)
```

```python

# Cell 8: Train Model

pipeline.fit(X_train, y_train)

print("=" * 50)
print("MODEL TRAINING COMPLETED SUCCESSFULLY")
print("=" * 50)
```

```python
# ==========================================================
# Cell 9: Model Evaluation
# ==========================================================

from sklearn.metrics import (
    accuracy_score,
    classification_report,
    confusion_matrix,
    roc_auc_score
)

# Predictions
y_pred = pipeline.predict(X_test)
y_prob = pipeline.predict_proba(X_test)[:, 1]

# Metrics
accuracy = accuracy_score(y_test, y_pred)
auc = roc_auc_score(y_test, y_prob)

print("=" * 50)
print("MODEL PERFORMANCE")
print("=" * 50)

print(f"Accuracy : {accuracy:.4f}")
print(f"ROC-AUC  : {auc:.4f}")

print("\nClassification Report")
print(classification_report(y_test, y_pred))

print("\nConfusion Matrix")
print(confusion_matrix(y_test, y_pred))
```

```python
# ==========================================================
# Save Model
# ==========================================================

import joblib

joblib.dump(
    pipeline,
    "donor_prediction_model.pkl"
)

print("Model Saved Successfully")
```
